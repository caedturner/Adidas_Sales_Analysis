```python
# Import Dependancies
import pandas as pd
```

```python
# Read CSV 
df = pd.read_csv('/Users/caed/Desktop/Post/Adidas/Adidas US Sales Datasets.csv')
```

```python
df
```

```python
# Drop null values
df.dropna()
```

```python
# Make row 3 the header of the DatFrame
header_row = df.iloc[3]
```

```python
df2 = pd.DataFrame(df.values[4:], columns=header_row)
df2
```

```python
# Check Data Types
df2.dtypes
```

```python
# Change "Invoice Date" to datetime
pd.to_datetime(df2['Invoice Date'])
```

```python
# Change numeric objects to float
df2['Retailer ID'] = df2['Retailer ID'].astype(float)
```

```python
# Check
df2.dtypes
```

```python
# use .replace() to remove unwanted characters - '$' ',' '%' 
df2['Price per Unit'] = df2['Price per Unit'].replace({'\$':''}, regex=True)
df2['Units Sold'] = df2['Units Sold'].replace({',':''}, regex=True)
df2['Total Sales'] = df2['Total Sales'].replace({'\$':'', ',':''}, regex=True)
df2['Operating Profit'] = df2['Operating Profit'].replace({'\$':'', ',':''}, regex=True)
df2['Operating Margin'] = df2['Operating Margin'].replace({'\%':''}, regex=True)
df2
```

```python
# Continue to change the numeric columns from objects to float 
df2['Price per Unit'] = df2['Price per Unit'].astype(float)
df2['Units Sold'] = df2['Units Sold'].astype(float)
df2['Total Sales'] = df2['Total Sales'].astype(float)
df2['Operating Profit'] = df2['Operating Profit'].astype(float)
df2['Operating Margin'] = df2['Operating Margin'].astype(float)
df2.dtypes
```
